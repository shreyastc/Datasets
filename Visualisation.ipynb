```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census=pd.read_csv('A:/AI/census_data.csv',sep=',') 
```

```python
census
```

```python
census.head(6)
```

```python
census.tail(7)
```

```python
plt.plot(census['age'], census['capital_gain'])
```

```python
plt.scatter(census['age'], census['capital_gain'])
```

```python
sns.set() # to get grey background
plt.scatter(census['age'], census['capital_gain'], s=15,c='red',marker='1')
plt.show()
```

```python
#s=size,c=color,marker=indicator
```

```python
plt.figure(figsize=(12,10)) #to adjust size of the plot area
plt.plot(census['capital_loss'],census['capital_gain'],color='darkviolet')
plt.show()
```

```python
plt.figure(figsize=(12,10)) #to adjust size of the plot area
sns.scatterplot(x='capital_loss',y='capital_gain',data=census)
plt.show()
```
